# 06 — Processed FPD image CSV import

This notebook demonstrates the processed FPD image export reader.

The processed FPD image file contains:

- metadata
- ROI / peak list
- embedded spectrum
- one or more image channels, such as `Total`

```python
%load_ext autoreload
%autoreload 2

from pathlib import Path

from pymagsims.image import SIMSImage
from pymagsims.plotting import plot_ion_image_grid, plot_ion_image

DATA = Path("../data")
img = SIMSImage.from_fpd_csv(DATA / "FPD_image2.csv")
```

## Metadata and channels

```python
display(img.metadata)
display(img.roi_table)
print("Image labels:", img.labels())
print("Total image shape:", img.total().shape)
```

## Embedded spectrum

```python
display(img.spectrum.data.head())
img.spectrum.plot(log_y=True);
```

## Plot processed image channels

```python
plot_ion_image_grid(img.images, log=True, ncols=2);
```
